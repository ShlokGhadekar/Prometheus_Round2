```python
import zipfile
import os

# Path to the uploaded zip file in Colab
zip_file_path = '/content/Ball Detection.v3i.yolov8.zip'  # This path may change depending on what you uploaded
extracted_folder = '/content/'

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)

# Verify the contents of the extracted folder
os.listdir(extracted_folder)
```

```python
# Check the contents of the train and valid directories
train_dir = '/content/train'
valid_dir = '/content/valid'

# List the files in the train directory
train_files = os.listdir(train_dir)
valid_files = os.listdir(valid_dir)

print("Training files:", train_files)
print("Validation files:", valid_files)
```

```python
from ultralytics import YOLO

# Load the YOLOv8 model (you can choose a pre-trained model or start from scratch)
model = YOLO('yolov8n.pt')  # or use yolov8x.pt for a more powerful model

# Train the model with your dataset
model.train(data='/content/data.yaml', epochs=50, imgsz=640)
```

```python
# Evaluate the model on validation data
results = model.val()
```

```python
from google.colab import files

# Upload the image
uploaded = files.upload()

# After uploading, you can use the filename to pass it to the model
img_path = list(uploaded.keys())[0]
```

```python
import matplotlib.pyplot as plt
import cv2

# Inference on the uploaded image
predictions = model(img_path)

# Get the results
img = cv2.imread(img_path)  # Read the image using OpenCV
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for matplotlib

# Draw bounding boxes on the image
for prediction in predictions:
    for box in prediction.boxes:
        x1, y1, x2, y2 = box.xyxy[0]  # Get coordinates of the bounding box
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)  # Draw the rectangle

# Display the image
plt.imshow(img)
plt.axis('off')  # Turn off axis
plt.show()
```
